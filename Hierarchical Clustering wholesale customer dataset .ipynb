```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
df = pd.read_csv('Wholesale customers data.csv')
df.head()
```

```python
from sklearn.preprocessing import normalize
df_scaled = normalize(df)
df_scaled = pd.DataFrame(df_scaled, columns=df.columns)
df_scaled.head()
```

```python
import scipy.cluster.hierarchy as sch 
plt.figure(figsize=(10, 7))
plt.title("Dendrograms")
dend = sch.dendrogram(sch.linkage(df_scaled, method='ward'))
```

```python
plt.figure(figsize=(10, 7))
plt.title("Dendrograms")
dend = sch.dendrogram(sch.linkage(df_scaled, method='ward'))
plt.axhline(y=6, color='red', linestyle='dashdot')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
cluster.fit_predict(df_scaled)
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(df_scaled["Milk"], df_scaled['Grocery'], c=cluster.labels_)
plt.title("Hierarchical Clustering for Milk and Grocery foods") 
```
