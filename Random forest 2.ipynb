```python
import pandas as pd 
import numpy as np 
from sklearn.ensemble import RandomForestClassifier
np.random.seed(0)
```

```python
db = pd.read_csv("diabetes.csv")
db.head()
```

```python
db["is_train"] = np.random.uniform(0, 1, len(db)) <=.8
db.head()
```

```python
train, test = db[db["is_train"]==True], db[db["is_train"]==False]
train
```

```python
test
```

```python
features = db.columns[:8]
features
```

```python
y = pd.factorize(train["Outcome"])[0]
y
```

```python
clf = RandomForestClassifier(n_estimators=100, n_jobs=2, random_state=0)
clf.fit(train[features], y)
```

```python
clf.predict(test[features])
```
